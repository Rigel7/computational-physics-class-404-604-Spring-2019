# Fluid Dynamics - Viscous and Incompressible 

The real fluid always has molecular viscosity. The Euler equation becomes the Navier-Stokes equation.

\begin{eqnarray}
\frac{D\rho}{Dt}&=&-\rho\nabla\cdot {\bf v}\\
\frac{D {\bf v}}{Dt}&=&\nu \nabla^2 {\bf v}-\frac{1}{\rho}\nabla P(\rho, T, x)
\end{eqnarray}
where $\nu$ is called the kinematic viscosity. 
 
Where the hydrodynamic derivative is defined as
\begin{eqnarray}
\frac{D\rho}{Dt}&=&\frac{\partial \rho}{\partial t}+({\bf v}\cdot \nabla)\rho\\
\frac{D {\bf v}}{Dt}&=&\frac{\partial {\bf v}}{\partial t}+({\bf v}\cdot {\bf \nabla}){\bf v}
\end{eqnarray}
The derivative is the rate of change in an element of flowing fluid. 
For the velocity, it introduces nonlinearities which eventually lead to all the interesting phenomena (e.g. turbulence). 

To simplify our problem, we will assume that
the fluid is incompressible. So the density $\rho$ is a constant and
\begin{equation}
\nabla\cdot{\bf v}=0
\end{equation}
This is also known as the condition of incompressibility.

Furthermore, we assume that the solution is steady. Then we have
\begin{equation}
({\bf v}\cdot\nabla){\bf v}=\nu\nabla^2{\bf v}-\frac{1}{\rho}\nabla P
\end{equation}

Thus, today we are solving these two equations
\begin{eqnarray}
\frac{\partial v_{x}}{\partial x}+\frac{\partial v_{y}}{\partial y}&=&0\\
\nu\left(\frac{\partial^2 v_{x}}{\partial x^2}+\frac{\partial^2 v_{x}}{\partial y^2}\right)&=&v_{x}\frac{\partial v_{x}}{\partial x}+v_{y}\frac{\partial v_{x}}{\partial y}+\frac{1}{\rho}\frac{\partial P}{\partial x}\\
\nu\left(\frac{\partial^2 v_{y}}{\partial x^2}+\frac{\partial^2 v_{y}}{\partial y^2}\right)&=&v_{x}\frac{\partial v_{y}}{\partial x}+v_{y}\frac{\partial v_{y}}{\partial y}+\frac{1}{\rho}\frac{\partial P}{\partial y}\\
\end{eqnarray}


## 2-D flow 

For incompressible flow, we can use stream function 
\begin{equation}
{\bf v}\equiv \nabla\times {\bf u}=\epsilon_{x}\left(\frac{\partial u_{z}}{\partial y}-\frac{\partial u_{y}}{\partial z}\right)+\epsilon_{y}\left(\frac{\partial u_{x}}{\partial z}-\frac{\partial u_{z}}{\partial x}\right)
\end{equation}
Here we assume that ${\bf v}$ only has two components.
In this way, we can use ${\bf u}$ to derive ${\bf v}$ which automatically satisfy
\begin{equation}
\nabla\cdot(\nabla\times{\bf u})=0
\end{equation}
Since $v_{x}$ and $v_{y}$ donot depend on $z$, ${\bf u}$ do not have z dependence, and we can write
\begin{equation}
{\bf v}=\epsilon_{x}\left(\frac{\partial u_{z}}{\partial y}\right)+\epsilon_{y}\left(-\frac{\partial u_{z}}{\partial x}\right)
\end{equation}
The reason that u is called the stream function is because contour lines u=constant are the streamlines. 

If we plug v into the equations, we will have third order differential equations. Instead we define another quantity
called vorticity
\begin{equation}
{\bf \omega}=\nabla\times {\bf v}=\nabla(\nabla\cdot {\bf u})-\nabla^2 {\bf u}
\end{equation}
For our 2-D problem, $\nabla\cdot {\bf u}$=0 so we have
\begin{equation}
\nabla^2{\bf u}=-{\bf \omega}
\end{equation}
If we use $\nabla\times$ for both sides of the full velocity equation
\begin{equation}
\frac{D {\bf v}}{Dt}=\nu \nabla^2 {\bf v}-\frac{1}{\rho}\nabla P(\rho, T, x)
\end{equation}
and use the vector identities
\begin{eqnarray}
({\bf v}\cdot\nabla){\bf v}&=&\nabla(\frac{1}{2}{\bf v}\cdot{\bf v})-{\bf v}\times\omega\\
\nabla\times({\bf v}\times\omega)&=&-\omega(\nabla\cdot{\bf v})+(\omega\cdot\nabla){\bf v}-({\bf v}\cdot\nabla)\omega\\
\nabla\cdot\omega&=&0\\
\nabla\times\nabla\phi&=&0
\end{eqnarray}
We have
\begin{equation}
\frac{D\omega}{Dt}=(\omega\cdot\nabla){\bf v}-\omega(\nabla\cdot{\bf v})+\frac{1}{\rho^2}\nabla\rho\times\nabla P+\nu\nabla^2\omega
\end{equation}
For incompressible fluid, we have
\begin{equation}
\frac{D\omega}{Dt}=(\omega\cdot\nabla){\bf v}+\nu\nabla^2\omega
\end{equation}
For our 2-D problem with steady state, we have
\begin{equation}
\nu\nabla^2\omega=[(\nabla\times {\bf u})\cdot\nabla]\omega
\end{equation}

So finally we have two non-linear elliptic PDEs that look like a mixture of Poisson's equation
with the wave equation. 
\begin{eqnarray}
\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}&=&-\omega\\
\nu\left(\frac{\partial^2 \omega}{\partial x^2}+\frac{\partial^2 \omega}{\partial y^2}\right)&=&\frac{\partial u}{\partial y}\frac{\partial \omega}{\partial x}-\frac{\partial u}{\partial x}\frac{\partial \omega}{\partial y}
\end{eqnarray}

## Finite differences form

If we make a gid of $N_{x}\times N_{y}$ of uniform spacing h, we have
\begin{equation}
\frac{\partial^2 u}{\partial x^2}+\frac{\partial^2 u}{\partial y^2}=\frac{u_{i+1,j}+u_{i-1,j}+u_{i,j+1}+u_{i,j-1}-4u_{i,j}}{h^2}
\end{equation}
So we have
\begin{eqnarray}
u_{i,j}&=&\frac{1}{4}(u_{i+1,j}+u_{i-1,j}+u_{i,j+1}+u_{i,j-1}+h^2\omega_{i,j})\\
\omega_{i,j}&=&\frac{1}{4}(\omega_{i+1,j}+\omega_{i-1,j}+\omega_{i,j+1}+\omega_{i,j-1})-\frac{R}{16}\left[(u_{i,j+1}-u_{i,j-1})*(\omega_{i+1,j}-\omega_{i-1,j})-(u_{i+1,j}-u_{i-1,j})*(\omega_{i,j+1}-\omega_{i,j-1})\right]
\end{eqnarray}
Since they are ellipitcal equations, we can use the successive over relaxation method
\begin{eqnarray}
u_{i,j}&+=&\Omega\left(\frac{1}{4}(u_{i+1,j}+u_{i-1,j}+u_{i,j+1}+u_{i,j-1}+h^2\omega_{i,j})-u_{i,j}\right)\\
\omega_{i,j}&+=&\Omega\left(\frac{1}{4}(\omega_{i+1,j}+\omega_{i-1,j}+\omega_{i,j+1}+\omega_{i,j-1})-\frac{R}{16}\left[(u_{i,j+1}-u_{i,j-1})*(\omega_{i+1,j}-\omega_{i-1,j})-(u_{i+1,j}-u_{i-1,j})*(\omega_{i,j+1}-\omega_{i,j-1})\right]-\omega_{i,j}\right)
\end{eqnarray}
